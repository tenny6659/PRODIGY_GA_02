```python
class Character:
    def __init__(self, name, role, traits,action):
        self.name = name
        self.role = role
        self.traits = traits
        self.action = action

    def describe(self):
        return f"{self.name} is a {self.role} with traits: {', '.join(self.traits)} and {self.action}."

# Example of creating a character
character = Character(name="Aria", role="Warrior",traits=["brave", "strong", "loyal"],action="fighting")
print(character.describe())
```

```python
class CharacterWithTasks(Character):
    def __init__(self, name, role, traits,action):
        super().__init__(name, role, traits,action)
        self.tasks = []

    def add_task(self, task):
        self.tasks.append(task)

    def perform_tasks(self):
        return [f"{self.name} is performing: {task}" for task in self.tasks]

# Create a character with tasks
character_with_tasks = CharacterWithTasks(name="Aria", role="Warrior", traits=["brave", "strong", "loyal"],action="fighting")
character_with_tasks.add_task("Defend the village")
character_with_tasks.add_task("Train with sword")
print(character.describe())
```

```python
results = character_with_tasks.perform_tasks()
for result in results:
    print(result)
```
